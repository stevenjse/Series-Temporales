# Cargar Datos

```python
import pandas as pd

# Ruta al archivo
ruta = 'dataset/stocks/TWTR.csv'

# Cargar el dataset
df = pd.read_csv(ruta)

# Convertir la columna de fecha a datetime
df['Date'] = pd.to_datetime(df['Date'])

# Establecer la columna de fecha como índice
df.set_index('Date', inplace=True)

# Establecer la frecuencia del índice
df = df.asfreq('D') # D: Diario, ME: Mensual, YE: Anual

# Comprobar valores nulos
print(df.isnull().sum())

# Relenar valores nulos si existen (interpolación)
if df.isnull().sum().sum() > 0:
    df.interpolate(method='linear', inplace=True) # linear, pchip...

    # Visualizar valores nulos
    print(df.isnull().sum())

# Visualizar el dataset
print(df.head())
```

# Random Walk

```python
import numpy as np

df['Return'] = np.log(df['Close'] / df['Close'].shift(1))
```

```python
np.random.seed(42)  # Para reproducibilidad

# Definir parámetros
num_steps = 1095  # Número de pasos
last_price = df['Close'].iloc[-1]

# Generar un paseo aleatorio
random_returns = np.random.normal(loc=df['Return'].mean(), scale=df['Return'].std(), size=num_steps)
random_walk = last_price * np.exp(np.cumsum(random_returns))

# Crear un DataFrame para el paseo aleatorio
dates = pd.date_range(start=df.index[-1], periods=num_steps)
random_walk_df = pd.DataFrame(data={'Close': random_walk}, index=dates)
```

```python
import matplotlib.pyplot as plt

# Graficar el precio de cierre y el paseo aleatorio
plt.figure(figsize=(12, 6))
plt.plot(df['Close'], label='Precio de Cierre')
plt.plot(random_walk_df['Close'], label='Paseo Aleatorio')
plt.title('Paseo Aleatorio Basado en Precios de Cierre')
plt.xlabel('Fecha')
plt.ylabel('Precio de Cierre')
plt.legend()
plt.show()
```

# Estacionariedad

```python
from statsmodels.tsa.stattools import adfuller

# Prueba de Dickey-Fuller
result = adfuller(random_walk_df['Close'].dropna())
print('Estadístico ADF:', result[0])
print('Valor p:', result[1])
print('Valores críticos:', result[4])

if result[1] <= 0.05:
    print('La serie es estacionaria')
elif result[0] > result[4]['5%']:
    print('La serie es no estacionaria')
else:
    print('La serie no es estacionaria')
```

# Estacionalidad

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomposición de la serie temporal
result = seasonal_decompose(random_walk_df['Close'].dropna(), model='additive')
result.plot()
plt.show()

result = seasonal_decompose(random_walk_df['Close'].dropna(), model='multiplicative')
result.plot()
plt.show()
```

# Autocorrelación

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

data = random_walk_df['Close'].dropna()

# Graficar ACF
plot_acf(data, lags=50, zero=False)
plt.title('Función de Autocorrelación (ACF)')
plt.show()

# Graficar PACF
plot_pacf(data, lags=50, zero=False, method='ols')
plt.title('Función de Autocorrelación Parcial (PACF)')
plt.show()
```
